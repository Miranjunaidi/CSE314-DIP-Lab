```python
img = []
y = 0
for i in bird:
    temp = []
    for j in i:
        #print(j)
        m = j[0]
        n = j[1]
        o = j[2]
        y = 0.2989 * m + 0.5870*n + 0.1140*o 
        y=y/181.018
        temp.append(y)
        #print(y)
    img.append(temp)
```
